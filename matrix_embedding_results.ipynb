```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


data = pd.read_csv('results/Bunka Paper - embeddigs.csv')

matrix_df = data.copy()
matrix_df = matrix_df.set_index('Unnamed: 0')
matrix_df = matrix_df.astype(float)
matrix_df = round(matrix_df, 2)
matrix_df.index.name = ""


# Plot
fig = sns.clustermap(
    matrix_df,
    cmap='Blues',
    annot=True,
    fmt=".2f",
    cbar_kws={"orientation": "horizontal", "label": None},  # Hide the color bar (legend)
    method='ward',
    figsize=(7, 7),  # Adjust the figure size
)
fig.cax.set_visible(False)
plt.ylabel('')
plt.xlabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('results/Bunka Paper - Sheet1.csv')
df = df.drop('Average ⬆️', axis=1)
df = df.set_index('Name')
df_t = df.transpose()



# Defining the color map for the models
color_map = {
    "Topic Neural Hermes": "blue", 
    "mlabonne/NeuralHermes-2.5-Mistral-7B": "orange", 
    "teknium/OpenHermes-2.5-Mistral-7B": "green"
}

# Reordering the columns to ensure that Topic Neural Hermes is displayed first
column_order = [
    "Topic Neural Hermes", 
    "mlabonne/NeuralHermes-2.5-Mistral-7B", 
    "teknium/OpenHermes-2.5-Mistral-7B"
]
df_t = df_t[column_order]

# Plotting the bar graph with tasks on the x-axis and models as the categories/colors
ax = df_t.plot(kind='bar', color=[color_map[model] for model in column_order], figsize=(10, 6))

# Adding the values on top of the bars
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height(), f'{i.get_height():.1f}', 
            ha='center', va='bottom', fontsize=8, rotation=0)


# Adding some customizations
plt.title('')
plt.xlabel('')
plt.ylabel('Performance Score')
plt.xticks(rotation=0)
plt.legend(title='Model')
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```
